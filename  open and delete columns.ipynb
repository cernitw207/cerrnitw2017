# Short tutorial on how to open and delete columns in a csv file

```python
# Import the pandas library and 'save' it as 'pd' : 

import pandas as pd
```

```python
# open the original file and save it as 'dataset':
dataset = pd.read_csv('C:\cern data/DoubleMuRun2011A.csv')
# Display the top of data:
dataset.head()
```

```python
f=pd.read_csv("C:\cern data/DoubleMuRun2011A.csv")

# pick columns that you would like to keep:


keep_col = ['px1','py1','pz1','px2', 'py2', 'pz2']
new_f = f[keep_col]

# saves the new csv with only momentum measurements to a new file:

new_f.to_csv("C:\cern data/newFile.csv", index=False)
```
